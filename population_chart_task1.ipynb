```python
import matplotlib.pyplot as plt

# Sample population data (in millions)
countries = ['India', 'China', 'USA', 'Indonesia', 'Pakistan',
             'Nigeria', 'Brazil', 'Bangladesh', 'Russia', 'Mexico']
population = [1440, 1420, 340, 277, 240, 230, 216, 170, 145, 130]

# Assign unique colors to each bar
colors = ['#FF9999','#66B3FF','#99FF99','#FFCC99','#C2C2F0',
          '#FFB266','#B3FF66','#FF66B2','#66FFE0','#D1B3FF']

# Create a bar chart
plt.figure(figsize=(12, 6))
bars = plt.bar(countries, population, color=colors, edgecolor='black', linewidth=1.5)

# Add value labels on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 10, f'{yval}', ha='center', va='bottom')

# Add titles and labels
plt.title('Top 10 Most Populous Countries (2024)', fontsize=16, color='darkblue')
plt.xlabel('Countries', fontsize=12)
plt.ylabel('Population (in millions)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()

# Show the bar chart
plt.show()
```
